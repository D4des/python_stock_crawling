```python
import FinanceDataReader as fdr

# 2010년부터 현재까지 삼성전자의 주식 정보 호출
df_samsung = fdr.DataReader('005930', '2010-01-01', '2023-02-05')

# 결과 출력
df_samsung.Close.plot()
```

```python
print(df_samsung)
```

```python
max_price = df_samsung['Close'].max()
max_data = df_samsung.loc[df_samsung['Close'] == max_price].index[0]
print("The highest price of Samsung Electronics was", max_price, "KRW on", max_data)
```

```python
import matplotlib.pyplot as plt

df_samsung['20d_MA'] = df_samsung['Close'].rolling(window=20).mean()

plt.plot(df_samsung.Close, label = 'Closing Price')
plt.plot(df_samsung['20d_MA'], label = '20-day Moving Average')
plt.xlabel('Year')
plt.ylabel('Price (KRW)')
plt.title('Closing Price and 20-day Moving Average of Samsung Electronics')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

df_samsung['20d_MA'] = df_samsung['Close'].rolling(window=20).mean()

plt.figure(figsize=(20, 15))
plt.plot(df_samsung.Close, label = 'Closing Price')
plt.plot(df_samsung['20d_MA'], label = '20-day Moving Average')
plt.xlabel('Year')
plt.ylabel('Price (KRW)')
plt.title('Closing Price and 20-day Moving Average of Samsung Electronics')
plt.legend()
plt.show()
```

```python
df_samsung.to_csv('samsung_electronics_stock_price.csv', index=False)
```

```python
df_samsung.to_csv('samsung_electronics_stock_price.csv', index=True)
```
